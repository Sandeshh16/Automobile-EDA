##INTRODUCTION
###Exploratory Data Analysis(EDA)
Exploratory Data Analysis (EDA) is a process in data science used to analyze datasets, summarize their key characteristics, and uncover patterns, anomalies, or relationships. It is essential for understanding the data before applying machine learning models or statistical techniques. EDA involves both numerical and graphical methods

EDA is crucial in data analysis projects! It acts as a detective, unveiling hidden patterns and relationships in the data. EDA helps in understanding the distribution and trends, acting like a compass guiding us through the data. It plays the role of a broom, cleaning the data by identifying missing values and errors. Like a puzzle solver, EDA helps select relevant features. It's a traffic signal, guiding data preparation decisions. EDA is an artist, visualizing data to communicate insights effectively. It fosters collaboration, encouraging discussions among team members. As a toolbox, it assists in model selection. EDA serves as a risk assessment tool, identifying potential challenges. Ultimately, it's a lightbulb, generating valuable insights and driving data-driven decision-making.

##Missing Values
As we can see we have missing values in our data, let's deal with the missing values

```python
#check missing values
data.isnull().sum()
```

```python
#drop the rows having null values in the num-of-doors field
data1 = data[~ data['num-of-doors'].isnull()]
```

Column bore has missing values, lets replace them with mean.

```python
data1['bore'].mean()
```

```python
#Fill the null values in bore column with mean
data1['bore'].fillna(data1['bore'].mean(), inplace = True)
```

```python
data1.isnull().sum()
```

```python
#Fill the null values in stroke column with mean
data1['stroke'].fillna(data1['stroke'].mean(), inplace = True)
```

```python
#Fill the null values in horsepower column with mean
data1['horsepower'].fillna(data1['horsepower'].mean(), inplace = True)
```

```python
#Fill the null values in peak-rpm column with mean
data1['peak-rpm'].fillna(data1['peak-rpm'].mean(), inplace = True)
```

```python
#check the nulls in the entire dataframe again
data1.isnull().sum()
```

##Handling incorrect data types


```python
data1.dtypes
```

```python
#analyse the num-of-doors column to check the issue
data1['num-of-doors']
```

```python
#write the function to change the data type of num-of-door column from categorical to numeric
def doors(value):
  return int(value.replace("two", "2").replace("four", "4"))
```

```python
data1['num-of-doors'] = data1['num-of-doors'].apply(doors)
data1['num-of-doors']
```

```python
#Analyse the num-of-cylinders column to check the issue
data1['num-of-cylinders']
```

```python
data1['num-of-cylinders'].value_counts()
```

```python
#write the function to change the data type of num-of-cylinders column from categorical to numeric
def cylinders(value):
  return int(value.replace('four','4').replace('six','6').replace('five','5').replace('eight','8').replace('two','2').replace('twelve','12').replace('three','3'))
```

```python
data1['num-of-cylinders'] = data1['num-of-cylinders'].apply(cylinders)
```

##**Univariate Analysis**
Univariate analysis focuses on the examination of a single variable in a dataset. It helps in understanding the distribution, central tendency, spread, and shape of the variable. This type of analysis does not consider relationships between variables but rather looks at each variable independently.

##Histogram
 Visualizes the frequency distribution of a numeric variable. Helps identify modes (peaks) in the data.

###**Histogram of Engine Size**

```python
sns.histplot(data1, x = 'engine-size', bins = 20, color = 'cyan', kde = True)
plt.title('Histogram of Engine Size\n')
plt.show()
```

### **Insights**
**Distribution:** The histogram reveals that the engine sizes are not uniformly distributed and are instead right-skewed.

**Common Range:** Most engine sizes fall between 100 and 200 units, indicating this is the typical range for cars in the dataset.

**Rare Large Engines:** Engine sizes greater than 300 are rare and likely belong to high-performance or specialized vehicles.

**Smooth Density Curve:** The KDE curve overlays the histogram and highlights the peak around 120-150 units.

##**Count plot of Car makers**

```python
data1['make'].value_counts()
```

```python
sns.countplot(data = data1, x = 'make', order = data1['make'].value_counts().index)
plt.title('Count Plot of Car Makers')
plt.xlabel('Car Makers')
plt.xticks(rotation = 90)
plt.show()
```

### **Insights**
**Toyota Leads:** Toyota is the most frequent make in the dataset, indicating its dominant presence, likely due to its global popularity and wide consumer base.

**Mid-Tier Representation:** Brands like Nissan, Mazda, and Mitsubishi are also prominent, reflecting their significant market share but slightly lower representation compared to Toyota.

**Rare Brands:** Makes like Alfa-Romeo, Isuzu, Renault, and Mercury have minimal representation, suggesting they have lower popularity in the regions covered by the dataset.

##**Bivariate Analysis**

Bivariate analysis examines the relationship between two variables to identify patterns, trends, or correlations. It helps us understand how one variable influences or is associated with another, providing deeper insights into data compared to univariate analysis.

###Scatter Plot between engine size and price

```python
sns.scatterplot(data=data1, x = 'engine-size', y = 'price', color = 'blue')
plt.title('Scatter plot of Engine size vs Price\n', color = 'green')
plt.xlabel('Engine Size', color = 'brown')
plt.ylabel('Price', color = 'brown')
plt.show()
```

###**Insights**
**Scatter Plot: Engine Size vs. Price**

**Positive Correlation:** Larger engine sizes generally correspond to higher prices, though the relationship is not perfectly linear.

**Clusters:** A dense cluster of points is observed for engine sizes between 100-150 units. Indicating that sales of affordable cars are higher.

###Bar Plot of Average price for each car make

```python
a = data1.groupby(by = 'make')['price'].mean()
a.sort_values(ascending = True, inplace = True)
a
```

```python
#Bar Plot : Average price for each car maker
plt.barh(a.index, a.values)
plt.title('Average price for each car maker\n', color = 'r')
plt.xlabel('Car makers', color = 'brown')
plt.ylabel('Average Price', color = 'brown')
plt.show()
```

###**Insights**
**Bar Plot: Average Price by Car Maker**

**High-End Brands:**Jaguar, Mercedes-Benz, and Porsche have the highest average prices, suggesting their vehicles cater to premium and luxury markets.

**Economical Brands:** Chevrolet, Dodge, and Plymouth are on the lower end, likely representing budget-friendly vehicles.

##Clustered Bar Plot: Comparing fuel-type and aspiration with respect to price

```python
# Clustered Bar Plot: Comparing fuel-type and aspiration with respect to average price
sns.barplot(data=data1, x='fuel-type', y='price', hue='aspiration')
plt.title('Clustered Bar Plot: Price by Fuel Type and Aspiration\n')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.show()
```

###**Insights**
**Clustered Bar Plot: Price by Fuel Type and Aspiration**

**Observation:** Turbocharged diesel cars have the highest prices, reflecting their efficiency and premium features.

**Gas-Powered Cars:** Standard aspiration gas cars have lower prices compared to their turbocharged cars

##Distribution of price across body style

```python
b = data1.groupby(by = 'body-style')['price'].mean()
b.sort_values(ascending = True, inplace = True)
b
```

```python
#Bar plot :Distribution of price across body styles
plt.barh(b.index, b.values, color = 'cyan', edgecolor = 'black')
plt.title('Distribution of price across body styles\n')
plt.xlabel('Average Price')
plt.ylabel('Body style')
plt.show()

```

###**Insights**
**Price Distribution by Body Style**

**Convertible and Hardtop Cars:** Have higher prices, indicating they might belong to premium categories.

**Hatchbacks:** Tend to be more affordable, with the lowest prices in given data.

##**Heatmap**
